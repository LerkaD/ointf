заданная функция : sin^2(x) - x/5


```python
a = 0 
b = 4
m = 11
n = 3
```

```python
xarr = []

for i in range(1,m):
    xarr.append(a + (i-1)*(b-a)/(m-1))

print (xarr)
```

```python
yarr = []

for x in xarr:
    yarr.append( math.sin(x) ** 2 - x/5)

print(yarr)
```

```python
plt.plot(xarr,yarr, 'o', linestyle = '-')
plt.show()
```

```python
newxarr = []

for j in range(1,21):
    newxarr.append(a + (j-1) * (b-a) / 20)

print(newxarr)
```

 sin^2(x) - x/5 от new_x_arr

```python
newyarr = []

for x in newxarr:
    newyarr.append( math.sin(x) ** 2 - x/5)

print(newyarr)
```

```python
plt.plot(newxarr,newyarr, 'o')
plt.show()
```

```python
fi_3 = interp1d(newxarr,newyarr, kind = 'quadratic')
yi_3 = fi_3(newxarr)
plt.plot(newxarr, yi_3, 'k:', label='квадратичная интерполяция', linestyle = '-', color = 'black') 
plt.plot(xarr,yarr, 'o',label=' график f(Xi)', linestyle = '-', color = 'red')
plt.plot(newxarr,newyarr, '*',label=' график f(Xj)',linestyle = '--', color = 'green')
plt.legend()
plt.show()
#print(yi_3)
```

Разница между f(Xj) и fi_3(Xj), где fi_3(Xj)- интерполир. функция

```python
delta = 0
delta_y = []
for index in range(len(newyarr)):
    delta += abs(newyarr[index] - yi_3[index])
    delta_y.append(abs(newyarr[index] - yi_3[index]))
print(f'Сумма отклонений: {delta}')
```

```python
plt.plot(newxarr, delta_y, label='график отклонениq',linestyle = '-', color = 'blue') 
plt.legend()
plt.show()
```

```python
fi_3 = interp1d(xarr,yarr, 'cubic')
yi_3 = fi_3(xarr)
plt.plot(xarr,yarr, 'o')
plt.plot(xarr, yi_3, 'k:', label='квадратичная интерполяция',linestyle = '-', color = 'black') 
plt.plot(xarr,yarr, 'o',label=' график f(Xi)')
plt.plot(newxarr,newyarr, '*',label=' график f(Xj)')
plt.legend()
plt.show()
```
